```python
import pandas as pd
import numpy as np

# this methods takes in a path to a csv file with the format input,output\n 
# and separates the inputs and outputs
#
# how to use:
# input, output = getInputOutput(path to csv)
def getInputOutput(path):
    data = pd.read_csv(path, header=None)
    return np.array(data[0]),np.array(data[1])


# sample csv
inp, out = getInputOutput("test.csv")

print(inp)
print(out)
```

```python
def getBenchmarkConstraints(path):
    file = open(path)

    text = file.readlines()
    constraints = []

    for i in range(len(text)):
        if "constraint" in text[i]:
            constraints.append(text[i][18:-3].split(")"))

    inputs = []
    outputs = []

    for i in range(len(constraints)):
        inputs.append(constraints[i][0].split("\"")[1])
        outputs.append(constraints[i][1].split("\"")[1])

    return inputs,outputs
        
```

```python
inputs, outputs = getBenchmarkConstraints("benchmarks/11440431.sl")

print("inputs: " + str(inputs))
print("outputs: " + str(outputs))
```

```python
import os

directory = 'benchmarks'

# iterate over files in directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    
    # skipping university files for now, will find a quick fix for that format later
    if os.path.isfile(f) and "univ" not in f:
        inputs, outputs = getBenchmarkConstraints(f)
        
    # printing a glimps of each file
    print("Inputs and Outputs of " + f)
    print("Inputs: " + str(inputs[0:3]) + "..." + str(inputs[-3:-1]))
    print("Outputs: " + str(outputs[0:3]) + "..." + str(outputs[-3:-1]))
    print()
```
